```python
import random
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x1 = [random.uniform(-1,1) for i in range(8000)]
x2 = [random.uniform(-1,1) for i in range(8000)]
```

```python
# plot train set
plt.plot(x1,x2,'ro')
```

```python
# train set
categ = []
for i in range(len(x1)):    
    if pow((x1[i]-0.5),2)+pow((x2[i]-0.5),2) < 0.2 and x2[i]>0.5:
        categ.append('1')
    elif pow((x1[i]-0.5),2)+pow((x2[i]-0.5),2) < 0.2 and x2[i]<0.5:
        categ.append('2')
    elif pow((x1[i]+0.5),2)+pow((x2[i]+0.5),2) < 0.2 and x2[i]>-0.5:
        categ.append('1')
    elif pow((x1[i]+0.5),2)+pow((x2[i]+0.5),2) < 0.2 and x2[i]<-0.5:
        categ.append('2')
    elif pow((x1[i]-0.5),2)+pow((x2[i]+0.5),2) < 0.2 and x2[i]>-0.5:
        categ.append('1')
    elif pow((x1[i]-0.5),2)+pow((x2[i]+0.5),2) < 0.2 and x2[i]<-0.5:
        categ.append('2')
    elif pow((x1[i]+0.5),2)+pow((x2[i]-0.5),2) < 0.2 and x2[i]>0.5:
        categ.append('1')
    elif pow((x1[i]+0.5),2)+pow((x2[i]-0.5),2) < 0.2 and x2[i]<0.5:
        categ.append('2')
    else:
        categ.append('3')
```

```python
len(categ)
```

```python
df = pd.DataFrame(list(zip(x1,x2,categ)),columns=['x1','x2','category'])
df.head() # tupwnoume ta 5 prwta data
```
